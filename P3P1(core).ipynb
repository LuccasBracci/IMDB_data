```python
import pandas as pd
import pymysql
import numpy as np
```

```python
basics_url= "https://datasets.imdbws.com/title.basics.tsv.gz"
ratings_url="https://datasets.imdbws.com/title.ratings.tsv.gz"
akas_url="https://datasets.imdbws.com/title.akas.tsv.gz"
```

```python
basics = pd.read_csv(basics_url, sep='\t', low_memory=False,)
ratings = pd.read_csv(ratings_url, sep='\t', low_memory=False,)
akas = pd.read_csv(akas_url, sep='\t', low_memory=False)
```

"Basics" Cleaning

```python
basics.replace({'\\N':np.nan}, inplace= True)
```

```python
basics['runtimeMinutes'].isna().sum()
```

```python
basics.dropna(subset=['runtimeMinutes','genres','startYear'], inplace=True)
```

```python
basics.info()
```

```python
basics['startYear']= basics['startYear'].astype(int)
```

```python
basics = basics[(basics['startYear']>=2000)&(basics['startYear']<2022)]
```

```python
basisc = basics.loc[basics['titleType'] == 'movie']

```

```python
basics
```

```python
basics = basics[basics['genres'].str.contains('Documentary')]
```

```python
basics
```

"Ratings" Cleaning

```python
ratings
```

```python
ratings.replace({'\\N':np.nan}, inplace= True)
```

```python
akas = akas.loc[(akas['region'] == 'US') & (akas['titleId'].isin(basics.loc[basics['titleType'] == 'movie']['tconst']))]
```

"akas" Cleaning

```python
akas
```

```python
akas.replace({'\\N':np.nan}, inplace= True)
```

```python
akas = akas.loc[akas['region'] == 'US']
```
